##  Analysis & Visualisation.

#### 1.Using line plot.

```python
def plot_energy_consumption(Renewable_Energy):
    """
    Plot energy consumption trends over the years for different sources.

    """
    """Extracting the  relevant columns"""
    
    # Extracting the 'Year' column
    years = Renewable_Energy['Year'] 
    
    # Extracting the 'Hydroelectric power' column.
    hydroelectric = Renewable_Energy['Hydroelectric power']
    
    # Extracting the 'Wind, wave, tidal' column
    wind_wave_tidal = Renewable_Energy['Wind, wave, tidal']
    
    # Extracting the 'Solar photovoltaic' column
    solar_pv = Renewable_Energy['Solar photovoltaic']
    
    # Extracting the 'Geothermal aquifers' column
    geothermal = Renewable_Energy['Geothermal aquifers']
    
    # Extracting the 'Landfill gas' column
    landfill_gas = Renewable_Energy['Landfill gas']
    
    # Extracting the 'Sewage gas' column
    sewage_gas = Renewable_Energy['Sewage gas']
    

    """ Plotting the enegry consumption over the years."""
    
    # Setting the figure size
    plt.figure(figsize=(12, 6))
    
    # Plotting 'Hydroelectric' data and adding the color 'blue'.
    plt.plot(years, hydroelectric, label='Hydroelectric', color='blue')
    
    # Plotting 'Wind, Wave, Tidal' data and adding the color 'green'.
    plt.plot(years, wind_wave_tidal, label='Wind, Wave, Tidal', color='green')
    
    # Plotting 'Solar PV' data and  adding the color 'orange'.
    plt.plot(years, solar_pv, label='Solar PV', color='orange')
    
    # Plotting 'Geothermal' data and adding the color 'red'.
    plt.plot(years, geothermal, label='Geothermal', color='red')
    
    # Plotting 'Landfill Gas' data and the color 'purple'.
    plt.plot(years, landfill_gas, label='Landfill Gas', color='purple')
    
    # Plotting 'Sewage Gas' data and the color 'brown'.
    plt.plot(years, sewage_gas, label='Sewage Gas', color='brown')
    
    # Adding title to the plot with font size '16'
    plt.title('Energy Consumption over the Years', fontsize=16)
    
    # Setting the x-axis label with font size '12'
    plt.xlabel('Year', fontsize=12)
    
    # Setting the y-axis label with font size '12'
    plt.ylabel('Energy Consumption (TWh)', fontsize=12)
    
    # Showing legend
    plt.legend(fontsize=10)
    
    # Showing grid
    plt.grid(True)
    
    # Rotating x-axis labels for better visibility
    plt.xticks(rotation=45)
    
    # Adjusting layout
    plt.tight_layout()
    
    
    # Showing the plot
    plt.show()
```

```python
"""
Calling the plot_energy_consumption function with the given data.

"""

plot_energy_consumption(Renewable_Energy)
```

```python
def plot_renewable_energy_production(Renewable_Energy):
    """
    Plotting a line plot for renewable energy production over the years.
    
    """
    # Setting the figure size
    plt.figure(figsize=(12, 6))
    
    # Plottng the line plot
    plt.plot(Renewable_Energy['Year'], Renewable_Energy['Energy from renewable & waste sources'], marker='o', linestyle='-', color='green')
    
    # Addding title to the plot with font size '16'
    plt.title('Renewable Energy Production in the UK (1990-2019)', fontsize=16)
    
    # Labelling the x-axis with font size '12'
    plt.xlabel('Year', fontsize=12)
    
    # Labelling the y-axis with font size '12'
    plt.ylabel('Renewable Energy Production (TWh)', fontsize=12)
    
    # Showing grid on the plot
    plt.grid(True)
    
    
    # Displaying the plot
    plt.show()
```

```python
"""
Calling  the plot function with the data

"""
plot_renewable_energy_production(Renewable_Energy)
```

```python
def plot_annual_percentage_change(Renewable_Energy):
    """
    Plot a line plot for the annual percentage change in consumption for each renewable energy source.

    """
    # Calculating annual percentage change in consumption for each renewable energy source
    Renewable_Energy['Hydroelectric_pct_change'] = Renewable_Energy['Hydroelectric power'].pct_change() * 100
    Renewable_Energy['Wind_pct_change'] = Renewable_Energy['Wind, wave, tidal'].pct_change() * 100
    Renewable_Energy['Solar_pct_change'] = Renewable_Energy['Solar photovoltaic'].pct_change() * 100

    # Setting the figure size
    plt.figure(figsize=(10, 6))

    # Plotting Hydroelectric Power data,adding linestyle and setting the marker.
    plt.plot(Renewable_Energy['Year'], Renewable_Energy['Hydroelectric_pct_change'], label='Hydroelectric Power', linestyle='-', marker='o')

    # Plotting Wind, Wave, Tidal data,adding linestyle and setting the marker
    plt.plot(Renewable_Energy['Year'], Renewable_Energy['Wind_pct_change'], label='Wind, Wave, Tidal', linestyle='-', marker='o')

    # Plotting Solar Photovoltaic data,adding linestyle and setting the marker
    plt.plot(Renewable_Energy['Year'], Renewable_Energy['Solar_pct_change'], label='Solar Photovoltaic', linestyle='-', marker='o')

    # Adding title to the plot and setting the fontsize to '16'
    plt.title('Annual Percentage Change in Renewable Energy Consumption', fontsize=16)

    # Labelling x-axis
    plt.xlabel('Year', fontsize=12)

    # Labelling y-axis
    plt.ylabel('Percentage Change', fontsize=12)

    # Showing legend
    plt.legend(fontsize=10)

    # Showing grid lines
    plt.grid(True)

    # Displaying the plot
    plt.show()
```

```python
"""
Call the function to plot annual percentage change

"""
plot_annual_percentage_change(Renewable_Energy)
```

#### 2.Using scatter plot.

```python
def plot_relationship_between_Renewable_Energy_Production_and_Total_Energy_Consumption(Renewable_Energy):
    """
    Creating a scatter plot with hue based on year.

    """
    # Setting the fig size.
    plt.figure(figsize=(10, 6))
    
    
    # Using seaborn to create a scatter plot with hue based on year
    sns.scatterplot(data=Renewable_Energy, x='Total energy consumption of primary fuels and equivalents', 
                    y='Energy from renewable & waste sources', hue='Year')
    
    # Adding title to the plot with font size 16
    plt.title('Relationship between Renewable Energy Production and Total Energy Consumption', fontsize=16)

    # Labelling x-axis with font size 12
    plt.xlabel('Total Energy Consumption (TWh)', fontsize=12)

    # Labelling y-axis with font size 12
    plt.ylabel('Renewable Energy Production (TWh)', fontsize=12)

    # Adding legend with title and position
    plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left',fontsize=10)

    
    # Showing grid lines
    plt.grid(True)
    
    # Displaying the plot
    plt.show()
```

```python
"""
Call the function to create scatter plot

"""
plot_relationship_between_Renewable_Energy_Production_and_Total_Energy_Consumption(Renewable_Energy)
```

#### 3. Using stacked area plot.

```python
def plot_renewable_sources(Renewable_Energy):
    """
     Plot a stacked area plot for renewable energy sources over the years.
    
    """
    # Selecting relevant columns for renewable energy sources
    renewable_sources = Renewable_Energy[['Year', 'Hydroelectric power', 'Wind, wave, tidal', 'Solar photovoltaic', 
                              'Geothermal aquifers', 'Landfill gas', 'Sewage gas']]

    # Plotting a stacked area plot for renewable energy sources over the years adding the fig size.
    renewable_sources.set_index('Year').plot(kind='area', stacked=True, figsize=(12, 6))

    # Adding title to the plot and setting the fontsize to '16'
    plt.title('Contribution of Renewable Energy Sources in the UK (1990-2019)',fontsize=16)

    # Labelling x-axis and setting the fontsize to '12'
    plt.xlabel('Year',fontsize=12)

    # Labelling y-axis and setting the fontsize to '12'
    plt.ylabel('Energy (TWh)',fontsize=12)

    # Showing grid lines
    plt.grid(True)

    # Displaying the plot
    plt.show()
```

```python
"""
Call the function to plot renewable energy sources'

"""
plot_renewable_sources(Renewable_Energy)
```

#### 4.Using density plot.

```python
def plot_renewable_density(Renewable_Energy):
    """
    Creating a density plot for renewable energy production.
    
    """
    # Setting the fig size
    plt.figure(figsize=(10, 6))
    
    # Using seaborn to create a histo plot with density distribution and color set as 'orange'.
    sns.histplot(Renewable_Energy['Energy from renewable & waste sources'], kde=True, color='orange')

    # Adding title to the plot with font
    plt.title('Density Distribution of Renewable Energy Production',fontsize=16)

    # Labelling x-axis
    plt.xlabel('Energy Production (TWh)',fontsize=12)

    # Labelling y-axis
    plt.ylabel('Density',fontsize=12)

    # Showing grid lines
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Adjustting layout
    plt.tight_layout()

    # Displaying the plot
    plt.show()
```

```python
"""
Call the function to plot the density distribution of renewable energy production

"""
plot_renewable_density(Renewable_Energy)
```

#### 5.Using grouped bar plot.

```python
def plot_renewable_energy_sources(Renewable_Energy):
    """
    Plot the energy production from various renewable sources over the years.

    """
    # Defining the energy sources
    energy_sources = ['Hydroelectric power', 'Wind, wave, tidal', 'Solar photovoltaic', 'Geothermal aquifers',
                      'Landfill gas', 'Sewage gas', 'Municipal solid waste (MSW)', 'Poultry litter', 'Straw',
                      'Wood', 'Charcoal', 'Liquid bio-fuels', 'Bioethanol', 'Biodiesel', 'Biomass']

    # Creating a figure with specified size
    plt.figure(figsize=(12, 8))

    # Iterate over each energy source to plot its data
    for energy_source in energy_sources:
        plt.bar(Renewable_Energy['Year'], Renewable_Energy[energy_source], label=energy_source)

    # Adding title to the plot with a fontsize of '16'
    plt.title('Energy from Various Renewable Sources (1990-2020)',fontsize=16)
    
    # Labelling the x-axis with a fontsize of '12'
    plt.xlabel('Year',fontsize=12)
    
    # Labelling the y-axis with a fontsize of '12'
    plt.ylabel('Energy (TWh)',fontsize=12)
    
    # Adding legend with customized position
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Showing the grid lines
    plt.grid(True)

    # Adjusting the layout
    plt.tight_layout()
    
    #Displaying the plot
    plt.show()
```

```python
"""
Call the function to plot the energy from various renewable sources

"""
plot_renewable_energy_sources(Renewable_Energy)
```

#### 6.Using heatmap.

```python
def plot_renewable_correlation(Renewable_Energy):
    """
    Plot a heatmap showing the correlation between different renewable energy sources.

    """
    # Selecting relevant columns for renewable energy sources
    renewable_sources = Renewable_Energy[['Hydroelectric power', 'Wind, wave, tidal', 'Solar photovoltaic', 
                               'Geothermal aquifers', 'Landfill gas', 'Sewage gas']]

    # Computing the correlation matrix
    correlation_matrix = renewable_sources.corr()

    # Creating a figure with specified size
    plt.figure(figsize=(10, 8))
    
    # Creating a heatmap using the seaborn function
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

    # Adding title to the plot
    plt.title('Correlation Between Renewable Energy Sources',fontsize=16)

    # Display the plot
    plt.show()
```

```python
"""
Call the function to plot the correlation heatmap

"""
plot_renewable_correlation(Renewable_Energy)
```
